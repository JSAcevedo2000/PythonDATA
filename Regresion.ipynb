```python
#Import libraries
import matplotlib.pyplot as plt
%matplotlib inline

#Data simulation
from sklearn.datasets import make_regression

#Simulation
x, y = make_regression(n_samples = 100, n_features = 1, noise = 10, bias = 50)

#Plotting the data
plt.plot(x, y, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Mean x and y
mean_x = sum(x) / len(y)
mean_y = sum(y) / len(x)

print("Mean x: ", mean_x)
print("Mean y: ", mean_y)
```

```python
#Initialization
m_numer = 0
m_denom = 0

#Computing m
for index in range (len(x)):
    m_numer += (x[index] - mean_x) * (y[index] - mean_y)
    m_denom += (x[index] - mean_x) ** 2
    
m = m_numer / m_denom

#Computing b
b = mean_y - m * mean_x

print("m = ", m)
print("b = ", b)
```

```python
#Plot real vs Prediction
y_model = m * x + b
```

```python
plt.plot(x, y, '.', label = "Scatter")
plt.plot(x, y_model, '-', label = "Linear Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
